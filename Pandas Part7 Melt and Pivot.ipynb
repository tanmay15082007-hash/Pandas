```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

## id_vars - what we dont want to change that is which column
## value_vars - which column do we want to melt
## var_name - Name of column under which the melted column will be present
## value_name - score where the integer part will be present

```python
df1 = df.melt(id_vars=["Name"],value_vars=["Math","Science","English"],var_name ="Subject",value_name="Score")
df1
```

```python
df1.pivot(index="Name",columns="Subject",values="Score")
```
