## Selecting Rows by Index
### Use .loc[] (label-based) and .iloc[] (position-based):

```python
df.loc[1]
```

```python
df.iloc[1]
```

```python
df.loc[0,'Actor'] # loc - by index
```

```python
df.iloc[0,0] # iloc - by position
```

```python
df.loc[0:2,["Actor","IMDb","Film"]]  # in loc last is inclusive as in comparision of this and below example
```

```python
df.iloc[0:2,0:2]  # in iloc last is not inclusive as 2 is not as compared to above loc example
```

```python
  df.at[0,"Actor"]  ## same function of loc but faster than loc as it is made for a specific element
```

```python
df.iat[0,0]   ## same function of loc but faster than loc as it is made for a specific element access
```

```python
df[df["IMDb"]>5]
```

```python
df[df["IMDb"]>5]['Actor']
```

```python
df[(df["IMDb"]>5)& (df['Year']>2019)]
```

```python
df[(df["IMDb"]>5) | (df['Year']>2019)]
```
