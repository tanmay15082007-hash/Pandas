```python
import pandas as pd
```

```python
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})

departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
```

```python
employees
```

```python
departments
```

```python
pd.merge(employees,departments, on="DeptID") # Ineer Join it skips the rows which are not common like in DEPTID it skipped 30 and 40 as they are not common in opposite of respective Dataframe
```

```python
pd.merge(employees,departments, on = "DeptID",how = "left") # LEft Join because it contains all of employees like DeptID 30 whose name element will be NaN
```

```python
pd.merge(employees,departments, on = "DeptID", how = "right")  # Right Join
```

```python
pd.merge(employees,departments,on = "DeptID", how = "outer") # Outer Join will show content of both dataframes and where data is missing it shows NaN
```
